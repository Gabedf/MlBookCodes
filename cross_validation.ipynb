```python
# FILTRO
value = int(input("Defina o valor: "))
filtro = df['quality'] == value
df[filtro].head(3)
```

```python
# MAPEANDO OS NÍVEIS DE QUALIDADE DE ACORDO COM O MAPEAMENTO
quality_mapping = {
    3 :0,
    4 :1,
    5 :2,
    6 :3,
    7 :4,
    8 :5 
    }
df.loc[:, "quality"] = df.quality.map(quality_mapping)
```
